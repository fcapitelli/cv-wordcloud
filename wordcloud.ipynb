# CV Wordcloud

```python
import os
from os import path
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
```

```python
# Read the whole text.
text = open(path.join(d, 'wordcloud_text.txt'), encoding='utf-8').read()
```

```python
# Stopwords list
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "vocÃª", "de", "ao", "os", "e", "para"])
```

```python
# Display the generated image:
wordcloud = WordCloud(width=1584, height=396, background_color='white', stopwords=stopwords).generate(text)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation='antialiased')
plt.axis("off")
```

```python
# lower max_font_size
wordcloud = WordCloud(width=1584, height=396, background_color='white', max_font_size=80, stopwords=stopwords).generate(text)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation="antialiased")
plt.axis("off")
plt.show()
```
